# DIMESIONALITY REDUCTION BY PRINCIPAL COMPONENT ANALYSIS(PCA)

The real life datasets have millions of features which needs lots of computation power and makes the training slower. This problem is referred to as "the curse of dimensionality". In order to overcome this, the PCA comes in as handy solution. It reduces the dimensions of the data by defining a hyperplane and then projects the two feature on that plane. for example, we have features x and y and we would like to define it by just one feature Z,PCA will define the hyperplane that is closest to the data and shows maximum variance.
PCA is used mainly for:
1.Data visualization
2.Increased efficiency 
3.reduction of requirement for memory.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data= pd.read_csv("iris.csv")
x=data.iloc[:,:4]
y=data.iloc[:,4]
```

```python
y
```

```python
data.columns=["sl","sw","pl","pw","species"] #Naming the columns
data.head()
```

```python
from sklearn.preprocessing import StandardScaler # Standardizing the data to have mean of 0 and std 1.
x = StandardScaler().fit_transform(x)
x
```

```python
data.corr() #The correlation matrix. 1 shows maximum correlation, The values close to 0 shows the two features are not dependent
            #on each other. the negative sign is to have inverse correlation.
```

```python
target=[]
for y in y:
    if y=="Iris-setosa":
        target.append(0)
    if y=="Iris-virginica":
        target.append(1)
    if y=="Iris-versicolor":
        target.append(2)
type(target)
target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)              #Here we would like to reduce the features from 4 to 2 principal components.
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
plt.scatter(principalDf['principal component 1'],principalDf['principal component 2'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y= train_test_split(principalDf,target,test_size= 0.3, train_size= 0.7 ,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(train_x,train_y)
```

```python
from sklearn.metrics import confusion_matrix
predictions = knn.predict(test_x)
cm = confusion_matrix(test_y,predictions)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_y,predictions)
accuracy
```
